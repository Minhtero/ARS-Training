## 1. C√†i ƒë·∫∑t & Import th∆∞ vi·ªán

```python

# N·∫øu ch∆∞a c√≥ th√¨ c√†i ƒë·∫∑t OpenCV (ch·ªâ c·∫ßn 1 l·∫ßn)
# !pip install opencv-python matplotlib

import cv2
import matplotlib.pyplot as plt

print("OpenCV version:", cv2.__version__)
```

## 2. ƒê·ªçc/ghi ·∫£nh

```python

# ƒê·ªçc ·∫£nh (v√≠ d·ª•: ·∫£nh m·∫∑c ƒë·ªãnh 'lena.png' ho·∫∑c ·∫£nh b·∫•t k·ª≥ em c√≥)
# ·ªû ƒë√¢y th·∫ßy s·∫Ω t·∫°o 1 ·∫£nh m·∫´u 128x128 ƒë·ªÉ minh h·ªça

import numpy as np

# T·∫°o ·∫£nh RGB gi·∫£ (m√†u ng·∫´u nhi√™n)
img = np.random.randint(0, 256, (128,128,3), dtype=np.uint8)

# L∆∞u ·∫£nh
cv2.imwrite("sample.png", img)

# ƒê·ªçc l·∫°i ·∫£nh
img2 = cv2.imread("sample.png")

# Hi·ªÉn th·ªã ·∫£nh v·ªõi matplotlib (v√¨ cv2.imshow kh√¥ng ch·∫°y ƒë∆∞·ª£c tr√™n Colab)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title("·∫¢nh g·ªëc")
plt.axis("off")
plt.show()
```

## 3. Resize ·∫£nh

```python

img_resized = cv2.resize(img2, (64,64))
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.title("·∫¢nh resize 64x64")
plt.axis("off")
plt.show()
```

## 4. Chuy·ªÉn ·∫£nh sang Grayscale

```python

gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.title("·∫¢nh Grayscale")
plt.axis("off")
plt.show()
```

## 5. Threshold nh·ªã ph√¢n

```python

# Threshold: gi√° tr·ªã > 127 = 255 (tr·∫Øng), ng∆∞·ª£c l·∫°i = 0 (ƒëen)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(binary, cmap="gray")
plt.title("·∫¢nh Threshold nh·ªã ph√¢n")
plt.axis("off")
plt.show()
```

## 6. L√†m m·ªù Gaussian Blur

```python

blur = cv2.GaussianBlur(gray, (5,5), 0)
plt.imshow(blur, cmap="gray")
plt.title("·∫¢nh l√†m m·ªù Gaussian Blur")
plt.axis("off")
plt.show()
```

## 7. B√†i t·∫≠p nh·ªè
